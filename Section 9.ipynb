# Categorical Plots

```python
tips = sns.load_dataset(name="tips")
tips.head()
```

```python
# average by default
sns.barplot(x="sex", y="total_bill", data=tips)
```

```python
# count
sns.barplot(x="sex", y="total_bill", data=tips, estimator=np.ma.count)
```

```python
sns.boxplot(x="day", y="total_bill", data=tips, hue="smoker")
```

# Grids

```python
iris = sns.load_dataset(name="iris")
iris.head()
```

```python
sns.pairplot(data=iris, vars=["sepal_length", "sepal_width", "petal_length", "petal_width"])
```

```python
g = sns.PairGrid(data=iris)
# g.map(plt.scatter)
g.map_diag(sns.histplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
```

```python
g = sns.FacetGrid(data=tips, row="smoker", col="time")
g.map(sns.histplot, "total_bill", kde=True)
```

```python
g = sns.FacetGrid(data=tips, row="smoker", col="time")
g.map(sns.scatterplot, "total_bill", "tip")
```

# Stacked Bar

```python
tips
```

```python
sns.barplot(data=tips, x="time", y="total_bill", estimator=np.sum,
            hue="sex", saturation=1)
# 饱和度 saturation
```

```python
sns.displot(data=tips, x="time", multiple="stack", hue="sex")
```

```python
# group = tips.groupby(by=["time", "sex"]).sum()
group = tips.groupby(by=["time", "sex"])
group
```

```python
group.aggregate(func={"tip":"sum"})
```

```python
group.aggregate(func={"tip":"sum"}).tip.unstack().plot.bar(stacked=True)
```

```python
group.aggregate(func={"tip":"sum"}).unstack().plot.bar(stacked=True)
```

```python
geyser = sns.load_dataset("geyser")
geyser
```

```python
sns.jointplot(x="waiting", y="duration", data=geyser)
```

# kind= "kde", "reg", "hist", "hex", "scatter" as default

```python
g = sns.jointplot(x=geyser["waiting"], y=geyser["duration"], kind="scatter")
g.plot_joint(sns.kdeplot, color="gray", level=5)
```

```python
g = sns.jointplot(x=geyser["waiting"], y=geyser["duration"], kind="kde", data=geyser,
                  hue="kind")
```

```python
g = sns.jointplot(x=geyser["waiting"], y=geyser["duration"], kind="scatter", data=geyser,
                  hue="kind", palette="autumn", height=8, space=0.2, ratio=10)
```

```python
g = sns.jointplot(x=geyser["waiting"], y=geyser["duration"], kind="reg",
                 joint_kws={"marker": "*", "order":2, "ci":True},
                 marginal_kws={"color": "gold"})
```
